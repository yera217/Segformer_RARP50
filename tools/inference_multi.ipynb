```python
import sys
import os

cur_dir = os.getcwd()
sys.path.append(os.path.dirname(cur_dir))

# Check Pytorch installation
import torch, torchvision
print(torch.__version__, torch.cuda.is_available())

# Check MMSegmentation installation
import mmseg
print(mmseg.__version__)
```

```python
from mmseg.apis import inference_segmentor, init_segmentor, show_result_pyplot
from mmseg.core.evaluation import get_palette

from mmseg.apis import inference_segmentor, init_segmentor, show_result_pyplot
from mmseg.core.evaluation import get_palette

config_file = '../work_dirs/yera.segformer/endvis_ce_160k/segformer.b1.512x512.rarp50.160k.py'
checkpoint_file = '../work_dirs/yera.segformer/endvis_ce_160k/iter_160000.pth'

model = init_segmentor(config_file, checkpoint_file, device='cuda:0')
```

## Make inference gif

```python
import glob
rgb_root_dir = '/media/deep/Transcend/sar-rarp-dataset/traindata/video_12.zip/rgb/'
save_dir = '../work_dirs/inference_vedio/inference/'

for i, rgb_name in enumerate(glob.glob(rgb_root_dir + '*.png')):
    print(i, rgb_name)
    result = inference_segmentor(model, rgb_name)
    out_filename = save_dir + rgb_name.split('/')[-1]
    model.show_result(rgb_name, result, wait_time=1, out_file=out_filename)
```

## Make label gif

```python
from PIL import Image
import numpy as np
import matplotlib as mpl

seg_root_dir = '/media/deep/Transcend/sar-rarp-dataset/traindata/video_12.zip/segmentation/'
save_dir = '../work_dirs/inference_vedio/label/'

for i, rgb_name in enumerate(glob.glob(rgb_root_dir + '*.png')):
    seg_name = seg_root_dir + rgb_name.split('/')[-1]
    img_seg = Image.open(seg_name)
                         
    img_seg = np.array(img_seg.convert('RGB'), dtype=int)
    img_seg = [img_seg[:,:,0]]
    out_filename = save_dir + rgb_name.split('/')[-1]
    model.show_result(rgb_name, img_seg, wait_time=1, out_file=out_filename)
    print(i, seg_name)
```

## Make inference video 60 fps

```python
import mmcv

# test a video and show the results
video = mmcv.VideoReader('/media/deep/Transcend/sar-rarp-dataset/traindata/video_12.zip/video_left.avi')
save_dir = '../work_dirs/inference_vedio/video/'
for i, frame in enumerate(video):
   result = inference_segmentor(model, frame)
   out_filename = save_dir + str(i) + '.png'
   model.show_result(frame, result, wait_time=1, show=True, out_file=out_filename)
```
